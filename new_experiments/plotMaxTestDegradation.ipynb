```python
# centers = [20,30,40,50,60,70,80,90,100,110,120,130,140,150,160,170,180,190,200,210,220,230,240,250,260]
# confs = {
#     "max":{c:configurations.config48[c] for c in centers},
#     "base":{c:configurations.config49[c] for c in centers},
# }

centers = [100,200,300,400,500,600,700,800,900,1000]
confs = {
    "MAX":{c:configurations.config50l[c] for c in centers},
    "ATT":{c:configurations.config64l[c] for c in centers},
    "MAXi":{c:configurations.config63l[c] for c in centers},
    "GRU":{c:configurations.config51l[c] for c in centers},
}
```

```python
maxes = {k:{c:fun.getMaxTest(confs[k][c]) for c in centers} for k in confs.keys()}
```

```python
plt.figure()
#plt.title("Maximum test accuracy")
for k in confs.keys():
    plt.plot(sorted(centers),[maxes[k][c] for c in sorted(centers)], label=k)
ax = plt.gca()
ax.minorticks_on()
ax.grid(which='major', linestyle='-', color='k')
ax.grid(which='minor', linestyle='--')
plt.xlabel(r"$m$")
plt.ylabel("accuracy")
plt.legend()
plt.show()
plt.savefig("tmp/maxBaseDiff.eps", bbox_inches='tight')
```
