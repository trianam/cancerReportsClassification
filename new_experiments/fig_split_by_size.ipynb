```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import pickle
plt.style.use(['ggplot'])
sns.set(rc={'figure.figsize':(11,8)})
##%%
df = pd.read_csv('./f1matrix-site.csv')
# df = df.sort_values(by='examples').reset_index()
classifiers = ['GRU','MAXi','MAXh','ATTh','ATT','MAX','SVM','XGBoost','BERT']
low = 5 # ignore below this
mid = 100 # upper bound of the easy class
high = 1000 # upper bount of the avg class
```

```python
t_sort.plot(ylim=(.88, .99),kind='line',x='examples',y=classifiers)
```
