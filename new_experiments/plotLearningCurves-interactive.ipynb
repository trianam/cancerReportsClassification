```python
loss = {k:fun.getLearningCurves(conf[k], 'loss') for k in conf}
acc = {k:fun.getLearningCurves(conf[k], 'accuracy') for k in conf}
```

```python
curves = loss
sets=['valid']

fig = plt.figure(figsize=(8,5))
fig.subplots_adjust(right=0.7)
ax = fig.add_subplot(1,1,1)

ax.set_xlabel('epoch')
ax.set_title("Loss")
lns = []
labels = []
for k in curves:
    for s in sets:
        lab = "{} - {}".format(k,s)
        ln, = ax.plot(curves[k][s], label=lab)
        lns.append(ln)
        labels.append(lab)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', color='k')
ax.grid(which='minor', linestyle='--')

interactive_legend = mpld3.plugins.InteractiveLegendPlugin(lns, labels)
mpld3.plugins.connect(fig, interactive_legend)
mpld3.display()
```

```python
curves = acc
sets=['valid']

fig = plt.figure(figsize=(15,30))
fig.subplots_adjust(right=0.7)
ax = fig.add_subplot(1,1,1)

ax.set_xlabel('epoch')
ax.set_title("Accuracy")
lns = []
labels = []
for k in curves:
    for s in sets:
        lab = "{} - {}".format(k,s)
        ln, = ax.plot(curves[k][s], label=lab)
        lns.append(ln)
        labels.append(lab)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', color='k')
ax.grid(which='minor', linestyle='--')

interactive_legend = mpld3.plugins.InteractiveLegendPlugin(lns, labels)
mpld3.plugins.connect(fig, interactive_legend)
mpld3.display()
```
