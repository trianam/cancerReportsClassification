```python
def plotCurves(curves, title="", sets=['train', 'valid', 'test']):
    plt.figure()
    plt.title(title)
    for k in curves:
        for s in sets:
            plt.plot(curves[k][s], label="{} - {}".format(k,s))
    ax = plt.gca()
    ax.minorticks_on()
    ax.grid(which='major', linestyle='-', color='k')
    ax.grid(which='minor', linestyle='--')
    plt.legend()
    plt.show()
```

```python
plotCurves(loss, "Loss", ['valid'])
```

```python
plotCurves(acc, "Accuracy",['valid'])
```
