```python
import math
import pandas as pd
from nltk import sent_tokenize
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktParameters
```

```python
corpusFolder = "./corpusLSTM_ICDO3/folds10"
textFileTrain = {}         
textFileTest = {}

lessEq100 = 0
lessEq200 = 0
lessEq300 = 0
lessEq400 = 0
maxLen = 0
avgLen = 0
sdLen = 0
lines = 0
tasks=['sede1', 'sede12', 'morfo1', 'morfo2']
for fold in range(10):
    print("Fold {}".format(fold))
    for task in tasks:
        print("    task {}".format(task))
        corpusFoldFolder = corpusFolder+"/"+task+"/"+str(fold)
        textFileTrain = corpusFoldFolder+"/textTrain.txt"
        textFileTest = corpusFoldFolder+"/textTest.txt"
        with open(textFileTrain) as fidTrain:
            with open(textFileTest) as fidTest:
                text = fidTrain.readlines() + fidTest.readlines()        
        
        for line in text:
            lines += 1
            words = len(line.split())
            if words > maxLen:
                maxLen = words
            avgLen += words
            sdLen += words * words
            
            if words <= 100:
                lessEq100 += 1
            if words <= 200:
                lessEq200 += 1
            if words <= 300:
                lessEq300 += 1
            if words <= 400:
                lessEq400 += 1
            
avgLen /= lines
sdLen = math.sqrt((sdLen/lines) - (avgLen*avgLen))
print("")
print("max len = {}".format(maxLen))
print("avg len = {} +/- {}".format(avgLen, sdLen))
print("less eq 100 = {:.1f}%".format((lessEq100/lines)*100))
print("less eq 200 = {:.1f}%".format((lessEq200/lines)*100))
print("less eq 300 = {:.1f}%".format((lessEq300/lines)*100))
print("less eq 400 = {:.1f}%".format((lessEq400/lines)*100))
```
