```python
import numpy as np
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open("./corpusLSTM_ICDO3/text.txt", 'r') as f:
    wc = Counter(" ".join(f.readlines()).strip().split())
```

```python
wc.most_common()
```

```python
x = []
y = []
c = [0]
for i,e in enumerate(wc.most_common()):
    #print(e)
    x.append(i+1)
    y.append(e[1])
    c.append(c[-1]+e[1])
c = c[1:]
p = list(map(lambda x: x/c[-1], c))
```

```python
plt.figure()
plt.plot(x,c)
plt.show()
```

```python
plt.figure()
plt.plot(x[:1000],c[:1000])
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca()
ax.set_xticks(np.arange(0,40000,5000))
ax.set_yticks(np.arange(0,1.05,0.05))
plt.plot(x,p)
plt.ylim([0.,1.05])
plt.savefig("/tmp/fig1.pdf", bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca()
ax.set_xticks(np.arange(0,1100,100))
ax.set_yticks(np.arange(0,1.05,0.05))
plt.plot(x[:1000],p[:1000])
plt.ylim([0.,1.05])
plt.savefig("/tmp/fig2.pdf", bbox_inches='tight')
plt.show()
```
